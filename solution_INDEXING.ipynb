```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url, sep='|')
```

Set 'user_id' as index

```python
users = users.set_index('user_id')
```

See first **rows**

```python
users.head()
```

Get 'age' column

```python
users['age']
```

Get multiple columns

```python
users[['age', 'gender', 'occupation']]
```

Get row by index (example user_id = 10)

```python
users.loc[10]
```

Get multiple rows

```python
users.loc[[1, 5, 10]]
```

Filter (age > 30)

```python
users[users['age'] > 30]
```

Filter by multiple conditions

```python
users[(users['age'] > 30) & (users['gender'] == 'M')]
```

# Select rows & columns together

```python
users.loc[1:10, ['age', 'occupation']]
```

iloc (position based)

```python
users.iloc[0:5]
```

Unique occupations

```python
users['occupation'].unique()
```

Count occupations

```python
users['occupation'].value_counts()
```

Sort by age

```python
users.sort_values(by='age')
```

BONUS

```python
users.loc[users['age'].idxmax()]
```
